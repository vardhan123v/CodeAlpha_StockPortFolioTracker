```python
import random

words = ["apple", "mango", "grape", "table", "tiger"]

word = random.choice(words)

guessed = []

chances = 6

while chances > 0:

    display = ""

    for letter in word:
        if letter in guessed:
            display = display + letter + " "
        else:
            display = display + "_ "

    print(display)

    if "_" not in display:
        print("You Won!")
        break

    guess = input("Enter a letter: ").lower()

    if guess in word:
        guessed.append(guess)
    else:
        chances = chances - 1
        print("Wrong Guess! Chances left:", chances)

if chances == 0:
    print("You Lost!")
    print("The word was:", word)
    
```
